```python
import nltk 
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
```

```python
swords = stopwords.words('english')
```

```python
def dataprep(sent):
    ps = PorterStemmer()
    swords = stopwords.words('english')
    lst_clean =[x for x in word_tokenize(sent) if x.lower() not in swords and x.isalnum()]
    return ([ps.stem(word) for word in lst_clean])
```

```python
x.apply(dataprep)
```

```python
# ## Spam dictionary
# lst_s = [key for key in (wc_s.words_).keys()]
# lst_n = [key for key in (wc_n.words_).keys()]
# print(lst_s)
```

## TF * IDF Vectorizer

```python
#import the class
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# create the object
tfidf = TfidfVectorizer(analyzer=dataprep)
```

```python
# apply
x_new = tfidf.fit_transform(x)
```

```python
# There are 6854 unique words in 5572 sentences
x_new
```

#### Cross Validation

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_new, y, random_state=0, stratify=y)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# By default the data is split in 75-25 ratio
1393/(4179+1393)
```

```python
### Build the machine learning classification model
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC(kernel='sigmoid', random_state=0)
```

```python
%%time
svm.fit(X_train, y_train)
```

```python
y_prob = svm.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, ConfusionMatrixDisplay
```

```python
accuracy_score(y_test, y_prob)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_prob)
```

```python
print(classification_report(y_test, y_prob))
```
