```python
alphabets='ABCDEFGHIJKLMNOPQRSTUVWXYZ'
```

```python
list(enumerate(alphabets))
```

```python
int_to_char=dict(enumerate(alphabets))
int_to_char
```

```python
#convert char to int
char_to_int={v:k for k,v in int_to_char.items()} #using dictionary comperhension
char_to_int
```

#### Data Preparation

```python
seq_length = 1
x = []
y = []
for i in range(len(alphabets)-seq_length):
  seq_in = [alphabets[i]]
  seq_out = alphabets[i+1]
  x += [[char_to_int[seq_in[0]]]]
  y += [char_to_int[seq_out]]
  print(seq_in, '->', seq_out)
```

Prepare the output variable

```python
from keras.utils import to_categorical
y_new = to_categorical(y)  ##one hot encoded format
y_new
```

```python
import numpy as np
```

```python
x = np.reshape(x, (25,1,1)) #total no. of words = 1 and length of words = 1
```

```python
x
```

```python
x.shape
```

```python
x_new = x/25.0 #Normalizing the data
x_new
```

Create the model

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
model = Sequential()

model.add(Dense(25, input_shape=(1,1), activation='relu'))
model.add(LSTM(32))
model.add(Dense(26, activation='softmax'))
```

```python
model.compile(loss= 'categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

Model training

```python
model.fit(x_new, y_new, epochs=1000, batch_size=1)
```

```python
# evaluate this model on test data, pass the whole data
loss, acc = model.evaluate(x_new, y_new, batch_size=1)
#print the loss and accuracy
print('Test Loss: %f' % (loss))
print('Test Accuracy: %f' % (acc))
```

Predict the character in sequence

```python
new = 'K'
new_int = [char_to_int[new]]
new_x = np.reshape(new_int, (1,1,1))
new_x = new_x/25.0
```

```python
#predict the next character
yhat = model.predict(new_x)
# int_to_char
int_to_char[np.argmax(yhat)]
```
