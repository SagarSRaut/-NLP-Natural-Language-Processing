## Data cleaning

```python
import re 
import string
def cleanResume(resumeText):
    resumeText = re.sub("http\S+\s*", " ", resumeText) #remove urls
    resumeText = re.sub("RT|cc", " ", resumeText) #remove RT cc
    resumeText = re.sub("#\S+", " ", resumeText) #remove hashtags
    resumeText = re.sub("@\S+", " ", resumeText) #remove mentions
    resumeText = re.sub("[%s]" % re.escape(string.punctuation), " ", resumeText) #remove 
    ## Remove punctuations
    resumeText = re.sub(r"[^\x00-\x7f]", r" ", resumeText)
    resumeText = re.sub("\s+", " ", resumeText) 
    ## Remove extra whitespace
    return resumeText
```

```python
#testing on one of the resume
cleanResume(x[128])
```

```python
cleaned = x.apply(cleanResume)
```

```python
cleaned
```

## TF * IDF Vectorizer

```python
#import the class
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# create the object
tfv = TfidfVectorizer(stop_words='english')
```

```python
# apply
x_new = tfv.fit_transform(cleaned)
```

```python
# There are 6854 unique words in 5572 sentences
x_new
```

```python
tfv.get_feature_names_out()
```

#### Cross Validation

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x_new, y, random_state=0, stratify=y)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
### Build the machine learning classification model
```

```python
from sklearn.neighbors import NearestCentroid
```

```python
nn = NearestCentroid()
```

```python
type(X_train)
```

```python
nn.fit(X_train, y_train)
```

## Performance Evaluation

```python
y_prob = nn.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, ConfusionMatrixDisplay
```

```python
accuracy_score(y_test, y_prob)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_prob)
```

```python
print(classification_report(y_test, y_prob))
```
