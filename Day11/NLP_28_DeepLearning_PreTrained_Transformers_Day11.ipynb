```python
from transformers import pipeline
```

```python
classifier = pipeline("text-classification")
```

```python
sent_analysis = pipeline("sentiment-analysis")
```

```python
import pandas as pd
text = 'It was possible for him easily. So he did it.'
result = classifier(text)
df = pd.DataFrame(result)
df
```

```python
text = 'tushar is dumb'
result = classifier(text)
pd.DataFrame(result)
```

```python
qa = pipeline('question-answering')
```

```python
text = 'Mumbai is capital city of Maharashtra'
q = 'What is the capital city of Maharashtra?'
```

```python
a = qa(question=q, context=text)
```

```python
a['answer']
```

Conversation

```python
from transformers import Conversation
conv = pipeline("conversational")
```

```python
conversation = Conversation("Hello, how are you?",model="microsoft/DialoGPT-small")
response = conv([conversation])
response[1]['content']
```

Visual question answering

```python
from PIL import Image
```

```python
vqa = pipeline('visual-question-answering')
```

```python
image = Image.open("car_damaged.jfif")

question = 'What is in the image?'
vqa(question=question, image=image, topk=3)
```

## Text to speech

```python
pipe = pipeline("text-to-speech", model="suno/bark-small")
```

```python
text = """Thou art the ruler of the minds of all people,
Thou Dispenser of India's destiny.
Thy name rouses the hearts of Punjab, Sind, Gujrat and Maratha,
Of Dravid, Orissa and Bengal.

It echoes in the hills of the Vindhyas and Himalayas,
Mingles in the music of Jamna and Ganges and is chanted by
the waves of the Indian sea.
They pray for Thy blessings and sing thy praise.
The saving of all people waits in thy hand,
Thou Dispenser of India's destiny,
Victory, Victory, Victory to Thee."""
pipe(text)
```

```python
from IPython.display import Audio
Audio(pipe(text)['audio'], rate=pipe(text)["sampling_rate"])
```
